```python
# 7.
import matplotlib.pyplot as plt

plt.bar(df['month_number'], df['total_profit'], color='blue')
# plt.ylim(100000, 500000)
plt.xlabel('Month number')
plt.ylabel('Total profit')
plt.title('Company profit per month')
plt.grid(True)
plt.show()
```

```python
# 8.
plt.plot(df['month_number'], df['total_profit'], linestyle='dotted', color='red', 
marker='o', markerfacecolor='black', linewidth=3, label='Profit data of last year')
plt.xticks(range(1, 13))
plt.ylim(100000, 500000)

plt.xlabel('Month number')
plt.ylabel('Total profit')
plt.title('Company profit per month')
plt.legend(loc='lower right')

plt.show()
```

```python
# 9.

df = df.rename(columns={'facecream': 'Face Cream Sales Data', 
                        'facewash': 'Face Wash Sales Data', 
                        'toothpaste': 'Tooth Paste Sales Data', 
                        'bathingsoap': 'Bathing Soap Sales Data', 
                        'shampoo': 'Shampoo Sales Data', 
                        'moisturizer': 'Moisturizer Sales Data'})

product = df[['Face Cream Sales Data', 'Face Wash Sales Data', 'Tooth Paste Sales Data', 
                   'Bathing Soap Sales Data', 'Shampoo Sales Data', 'Moisturizer Sales Data']]

product.plot(kind='line', marker='o')

plt.xticks(range(1, 13))
plt.ylim(1000, 18000)
plt.xlabel('Month Number')
plt.ylabel('Sales Unit in Number')
plt.title('Sales data')
plt.legend(loc='upper left')

plt.show()

```

```python
# Section 2
# 2.

height = np.array([121.9,124.5,129.5,134.6,139.7,147.3,152.4,157.5,162.6])
weight = np.array([19.7,21.3,23.5,25.9,28.5,32.1,35.7,39.6,43.2])

avg_hw = (height + weight) / 2

plt.plot(weight, avg_hw, linestyle='-.', color='green', marker='o', markersize=10,
         markerfacecolor='green', label='Avg weight vs height')

plt.xlabel('Weight in kg')
plt.ylabel('Height in cm')
plt.title('Average weight with respect to average height')
plt.legend(loc='lower right')

plt.show()
```
